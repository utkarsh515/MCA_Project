```python
# üìä Cell 1 (fixed): Explore dates & pick T‚ÇÄ as a Timestamp

import pandas as pd
import matplotlib.pyplot as plt

# 1) load & parse
df = pd.read_csv("OnlineRetail.csv", encoding="ISO-8859-1")
df = df.dropna(subset=["InvoiceDate"])
df["InvoiceDate"] = pd.to_datetime(
    df["InvoiceDate"],
    format="%d-%m-%y %H:%M",
    errors="coerce"
)

# 2) full Timestamp range
min_ts = df["InvoiceDate"].min()
max_ts = df["InvoiceDate"].max()
print(f"Timestamp range: {min_ts}  ‚Üí  {max_ts}")

# 3) monthly invoice counts
df.set_index("InvoiceDate").resample("ME")["InvoiceNo"].nunique().plot(figsize=(8,3))
plt.title("Unique Invoices per Month"); plt.ylabel("Count"); plt.show()

# 4) quantiles (for reference)
quantiles = df["InvoiceDate"].quantile([0.5,0.6,0.7,0.8,0.9])
print("Date quantiles:\n", quantiles)

# 5) pick T0 = max_ts ‚Äì 90 days (still a Timestamp)
T0 = max_ts - pd.Timedelta(days=90)
print(f"\n‚úîÔ∏è Selected cutoff T‚ÇÄ = {T0}")
```

```python
# üîß Cell 2: Build time-based RFM & true future churn label

import pandas as pd
from datetime import timedelta

# df from Cell 1 (has InvoiceDate as datetime64, etc.)
df = df.copy()
df["TotalPrice"] = df["Quantity"] * df["UnitPrice"]

# 1) feature vs label windows
feat_df  = df.loc[df["InvoiceDate"] <  T0].copy()
label_df = df.loc[
    (df["InvoiceDate"] >= T0) &
    (df["InvoiceDate"] <  T0 + timedelta(days=90))
].copy()

# 2) compute RFM on feature window
snapshot = feat_df["InvoiceDate"].max() + timedelta(days=1)
rfm = feat_df.groupby("CustomerID").agg({
    "InvoiceDate": lambda x: (snapshot - x.max()).days,
    "InvoiceNo":    "nunique",
    "TotalPrice":   "sum"
}).rename(columns={
    "InvoiceDate":"Recency",
    "InvoiceNo":"Frequency",
    "TotalPrice":"Monetary"
})

# 3) churn label from actual future purchases
p90 = (label_df.groupby("CustomerID")["InvoiceNo"]
        .nunique()
        .rename("Purchases90"))
rfm = rfm.join(p90, how="left").fillna(0)
rfm["Churn"] = (rfm["Purchases90"] == 0).astype(int)
rfm.drop(columns="Purchases90", inplace=True)

print("True churn distribution:\n", rfm["Churn"].value_counts(normalize=True))
rfm.head()
```

```python
# üß† Cell 3: Train/test, scale, up-sample, train & evaluate ANN

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.utils import resample
from sklearn.metrics import (
    accuracy_score, recall_score, f1_score,
    confusion_matrix, classification_report, roc_auc_score, roc_curve
)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
import seaborn as sns  # Import seaborn here
import matplotlib.pyplot as plt # Also ensure matplotlib is imported as it's used for plotting


import numpy as np
import tensorflow as tf
import random
import os

SEED = 33
np.random.seed(SEED)
tf.random.set_seed(SEED)
random.seed(SEED)
os.environ['PYTHONHASHSEED'] = str(SEED)


# features & target
X = rfm[["Recency","Frequency","Monetary"]]
y = rfm["Churn"]

# stratified split
x_train, x_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# scale
scaler = StandardScaler().fit(x_train)
x_train_s = scaler.transform(x_train)
x_test_s  = scaler.transform(x_test)

# up-sample minority in training set
train_df = pd.DataFrame(x_train_s, columns=X.columns)
train_df["Churn"] = y_train.values
maj = train_df[train_df["Churn"]==0]
minr= train_df[train_df["Churn"]==1]
min_up = resample(minr, replace=True, n_samples=len(maj), random_state=42)
bal   = pd.concat([maj,min_up]).sample(frac=1, random_state=42)
y_train_bal = bal["Churn"].values
x_train_bal = bal.drop("Churn", axis=1).values

# build & train ANN
model = Sequential([
    Dense(32, activation="relu", input_dim=3),
    Dropout(0.6), BatchNormalization(),
    Dense(64, activation="relu"), Dropout(0.6), BatchNormalization(),
    Dense(128, activation="relu"),Dropout(0.6),BatchNormalization(),
    Dense(1, activation="sigmoid")
])
model.compile(optimizer=Adam(1e-3), loss="binary_crossentropy", metrics=["accuracy"])
es = EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True)

history = model.fit(x_train_bal, y_train_bal,
                    validation_split=0.2, epochs=50,
                    batch_size=32, callbacks=[es], verbose=0)

# plots
plt.plot(history.history["accuracy"], label="train acc")
plt.plot(history.history["val_accuracy"], label="val acc")
plt.legend(); plt.title("Accuracy"); plt.show()
plt.plot(history.history["loss"], label="train loss")
plt.plot(history.history["val_loss"], label="val loss")
plt.legend(); plt.title("Loss"); plt.show()

# evaluate
y_prob = model.predict(x_test_s).flatten()
y_pred = (y_prob>0.5).astype(int)
print("Accuracy:", accuracy_score(y_test,y_pred))
print("Recall :", recall_score(y_test,y_pred))
print("F1     :", f1_score(y_test,y_pred))
print("ROC AUC:", roc_auc_score(y_test,y_prob))
print("\n"+classification_report(y_test,y_pred))

cm = confusion_matrix(y_test,y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["NotChurn","Churn"], yticklabels=["NotChurn","Churn"])
plt.title("Confusion Matrix"); plt.show()

fpr,tpr,_ = roc_curve(y_test,y_prob)
plt.plot(fpr,tpr,label=f"AUC={roc_auc_score(y_test,y_prob):.3f}")
plt.plot([0,1],[0,1],"--",color="gray")
plt.title("ROC Curve"); plt.legend(); plt.show()
```

```python
# üîç Cell 4: Dynamic predictions & comparison

# predict on all
X_all_s = scaler.transform(rfm[["Recency","Frequency","Monetary"]])
rfm["Churn_Pred"]    = (model.predict(X_all_s).flatten()>0.5).astype(int)
rfm["Churn_Prob(%)"] = (model.predict(X_all_s).flatten()*100).round(2)

# compare rule vs model churn
comp = rfm.groupby(["Churn","Churn_Pred"]).size().reset_index(name="Count")
print("Rule-vs-Model:\n", comp)
rfm[rfm["Churn"]!=rfm["Churn_Pred"]].head()
```
